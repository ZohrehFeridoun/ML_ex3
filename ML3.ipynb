```python
import sklearn
import numpy as np
import pandas as pd
import csv
from re import X

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn import metrics

from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_selection import RFE
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error
from sklearn import linear_model
from sklearn.preprocessing import Normalizer
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error

from sklearn.linear_model import SGDRegressor
from sklearn.model_selection import GridSearchCV
```

```python
data = pd.read_csv("/content/Advertising.csv")
data.head()
```

```python
data.drop('Unnamed: 0', axis = 1, inplace=True)
display(data)
```

```python
X = data[['TV', 'Radio', 'Newspaper']]
Y = data['Sales']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
transformer = Normalizer().fit(X_train)
normalized_X = transformer.transform( X_test)
print(normalized_X)
```

part 2

```python
model = LinearRegression()

model.fit(X_train, y_train)
```

```python
y2 = model.predict(X_test)

MSE = metrics.mean_squared_error(y_test, y2)

R2 = metrics.r2_score(y_test, y2)

print('MS:', MSE)
print('R:', R2)
```

```python
sns.scatterplot(y_test, y2)

plt.title('Linear Regression ')

plt.show()
```

```python
mse = mean_squared_error(y_test , y2)

error=pd.DataFrame({'Target':y_test,'Predicted':y2,'MSE':y_test - y2})

print(error)
```
